# 과일 개수 골라내기

> 장바구니에 아래와 같은 과일이 들어있고 과일 판별 리스트가 있습니다.

```python
basket_items = {'apples': 4, 'oranges': 19, 'kites': 3, 'sandwiches': 8}

fruits = ['apples', 'oranges', 'pears', 'peaches', 'grapes', 'bananas']

```
> 현재 장바구니에는 과일이 몇개이고 과일이 아닌 것은 몇개인지 출력하시오.
---
```
예시 출력)
과일은 23개이고, 11개는 과일이 아닙니다..
```

````python
basket_items = {'apples': 4, 'oranges': 19, 'kites': 3, 'sandwiches': 8}
fruits = ['apples', 'oranges', 'pears', 'peaches', 'grapes', 'bananas']
# 아래에 코드를 작성하세요.
````

````python
#dictionary count
#dictionary에서 key와 value를 모두 가져와서 key가 과일 리스트에 있으면 특정 카운트에 그 키의 값을 계속해서 더한다.
#반대로 키가 과일 리스트에 없으면, 다른 카운트에 그 키의 값을 계속해서 더한다.
fruit_count = 0
not_fruit_count = 0
for fruit, count in basket_items.items():
    if fruit in fruits:
        fruit_count += count
    else:
        not_fruit_count += count
print(f'과일은 {fruit_count}개이고, {not_fruit_count}개는 과일이 아닙니다.')
````

````python
gwail = 0
others = 0
for item in basket_items.keys():
    if item in fruits:
        gwail += basket_items.values()
    else:
        others += basket_items.values()
    

print(f'과일은 {gwail}개이고, {others}개는 과일이 아닙니다.')
````


````python
#이름을 받는다.
name = input() #type이 스트링
#리스트로 바꾼다.
names = name.split()
#반복문을 통해 처음과 끝 요소를 제외한 나머지 요소들을 '대문자+.'으로 변경한다.
for i in range(1, len(names)-1):
#     print(names[i])
    names[i] = names[i][0].upper() + '.'
print(names)
#리스트를 문자열로 바꾼다.  .join()
print(' '.join(names))#공백을 기준으로 names의 요소들을 합치겠다.
````

````python
calendar = { 
        1: 31, 2: 28, 3: 31, 4: 30, 5: 31,6: 30, 
        7: 31, 8: 31, 9: 30, 10: 31, 11: 30, 12: 31 
} 
weeks = ['Mo', 'Tu', 'We', 'Th', 'Fr', 'Sa', 'Su'] 
# 아래에 코드를 작성하세요.
````

````python
#딕셔너리의 키, 값을 반복문을 통해 이용한다.
#하나씩 해보면서 구조 맞춰나가기
for month, count_day in calendar.items():
    count = 0#월마다 초기화되어야 해서.. count는 일자가 한 줄에 7일만 출력되기 위한 변수
    print(f'{month:9} 월')
    for day in weeks:
        print(f'{day}', end = ' ')
    print()
    for i in range(1, count_day+1):
        print(f'{i:2}', end = ' ')
        count += 1
        if count == 7:
            print()
            count = 0
    print()
````


````python
# 아래에 코드를 작성하세요.
````

````python
phone = input()
if phone[0:3] == '010' and len(phone) == 11:
#     pass#나중에 개발하려고 일단 넘기는 것
    result = '*'*7 + phone[-4:]
    print(result)
else:
    print('핸드폰 번호를 입력해주세요.')
````

# 소수 찾기

> 조건, 반복문을 응용하여 numbers 리스트의 요소들이 소수인지 아닌지 판단하는 코드를 작성하라.

```python
numbers = [26, 39, 51, 53, 57, 79, 85]

# 출력 예시
26 는 소수가 아닙니다. 2 는 26 의 인수입니다.
39 는 소수가 아닙니다. 3 는 39 의 인수입니다.
51 는 소수가 아닙니다. 3 는 51 의 인수입니다.
53 는 소수입니다.
57 는 소수가 아닙니다. 3 는 57 의 인수입니다.
79 는 소수입니다.
85 는 소수가 아닙니다. 5 는 85 의 인수입니다.
```

````python
numbers = [26, 39, 51, 53, 57, 79, 85]
# 아래에 코드를 작성하세요.
````

````python
#반복문의 else를 사용하는 것이 포인트!
for number in numbers:
    for x in range(2, number):
        if number % x == 0:
            print(f'{number}는 소수가 아닙니다. {x}는 {number}의 인수입니다.')
            break
    else:
        print(f'{number}는 소수입니다.')
````
