# 최대공약수, 최소공배수 구하기

> 두 수를 입력받아 두 수의 최대공약수와 최소공배수를 반환하는 함수, gcdlcm을 완성해 보세요.
>
> 배열의 맨 앞에 최대공약수, 그다음 최소공배수를 넣어 반환하면 됩니다. 
>
> 예를 들어 두 수 3, 12의 최대공약수는 3, 최소공배수는 12이므로 `gcdlcm(3, 12)`는 `(3, 12)`를 반환해야 합니다. 

예시)

```python
print(gcdlcm(3, 12)) #=> (3, 12)
print(gcdlcm(1071, 1029)) #=> (21, 52479)
```

````python
# 여기에 코드를 작성하세요.
````

````python
def gcd(x, y):
    while y != 0:
        (x, y) = (y, x % y)
    return x

def lcm(x, y):
   if x > y:
       z = x
   else:
       z = y

   while(True):
       if((z % x == 0) and (z % y == 0)):
           lcm = z
           break
       z += 1

   return lcm
````

````python
def gcdlcm(x, y):
    a, b = max(x, y), min(x, y)
    while b != 0:
        (a, b) = (b, a % b)
    gcd = a
    lcm = int((x*y)/gcd)
    return(gcd, lcm)
    
````

````python
print(gcd(1029, 1071))
````

````python
print(gcdlcm(3, 12))
print(gcdlcm(1071, 1029))
````
