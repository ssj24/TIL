# 최대공약수, 최소공배수 구하기

> 두 수를 입력받아 두 수의 최대공약수와 최소공배수를 반환하는 함수, gcdlcm을 완성해 보세요.
>
> 배열의 맨 앞에 최대공약수, 그다음 최소공배수를 넣어 반환하면 됩니다. 
>
> 예를 들어 두 수 3, 12의 최대공약수는 3, 최소공배수는 12이므로 `gcdlcm(3, 12)`는 `(3, 12)`를 반환해야 합니다. 

예시)

```python
print(gcdlcm(3, 12)) #=> (3, 12)
print(gcdlcm(1071, 1029)) #=> (21, 52479)
```

````python
# 여기에 코드를 작성하세요.
````

````python
def gcd(x, y):
    while y != 0:
        (x, y) = (y, x % y)
    return x

def lcm(x, y):
   if x > y:
       z = x
   else:
       z = y

   while(True):
       if((z % x == 0) and (z % y == 0)):
           lcm = z
           break
       z += 1

   return lcm
````

````python
def gcdlcm(x, y):
    a, b = max(x, y), min(x, y)
    while b != 0:
        (a, b) = (b, a % b)
    gcd = a
    lcm = int((x*y)/gcd)
    return(gcd, lcm)
    
````

````python
#최소공배수 = 두 수의 곱 / 최대공약수
#둘 중의 큰 수를 찾아서 작은 수로 나눠나가는 것★
#더 이상 큰 수를 나눌 수 없을 때가 최대 공약수가 나옴

def gcdlcm(n, m):
    a, b = max(n, m), min(n, m)
    while b > 0:
        a, b = b, a%b
    return a, int(n*m/a)
````

````python
print(gcdlcm(3, 12))
print(gcdlcm(1071, 1029))
````

````python
#높이를 기준으로 낮시간 값은 높이 - 낮시간, 밤시간 값은 높이 + 밤시간
def snail(height, day, night):
    count = 0
    #남은 높이가 0이 되거나 0보다 작아지는 순간 반복문이 종료
    while True:#뭐가 true...? 걍 반복하겠다는 뜻인가봐..
        count += 1
        height -= day
        if height <= 0:
            return count
        height += night
````

# 무엇이 중복일까

> 다음 리스트에서 중복되는 요소만 뽑아서 새로운 리스트에 옮기시오. 

```python
# 입력)
duplicated(['a', 'b', 'c', 'b', 'd', 'm', 'n', 'n', 'b'])

# 출력)
['b', 'n']
```

````python
# 여기에 코드를 작성하시오.
````

````python
duplicated=['a', 'b', 'c', 'b', 'd', 'm', 'n', 'n', 'b']
````

````python
def duplicated(x):
    compare = [x[0]]
    result = []
    for i in x[1:]:
        if i in compare and i not in result:
            result.append(i)
        else:
            compare.append(i)
    return result
duplication=['a', 'b', 'c', 'b', 'd', 'm', 'n', 'n', 'b']
print(duplicated(duplication))
````

````python
#.count()
#중복 문제는 set이 좋다
def duplicated(words):
    duplicates = []
    for word in words:
        #만약에 현재 단어가 리스트에 2개 이상 있고, duplicates에 없다면 값을 추가해야한다.
        if words.count(word) > 1 and word not in duplicates:
            duplicates.append(word)
    return duplicates
duplication=['a', 'b', 'c', 'b', 'd', 'm', 'n', 'n', 'b']
print(duplicated(duplication))
````

````python
#set()
def duplicated(words):
    return list({word for word in words if words.count(word) > 1})
duplication=['a', 'b', 'c', 'b', 'd', 'm', 'n', 'n', 'b']
print(duplicated(duplication))
````

# 알파벳만 남기고 뒤집기

> 문자열이 주어지면, 해당 문자열 중에서 알파벳이 아닌 문자는 전부 빼고 거꾸로 뒤집어 반환하는 함수를 작성하세요.

```python
# 예시) 
reverse_letter('krishan')
'nahsirk'

reverse_letter('ultr53o?n')
'nortlu'
```

````python
# 여기에 코드를 작성하세요
````

````python
def reverse_letter(words):
    result = []
    for word in words:
        if word.isalpha():
            result.append(word)
    return ''.join(result[::-1])
````

````python
def reverse_letter(words):
    result = [word for word in words if word.isalpha()]
    return ''.join(result[::-1])
````

````python
def reverse_letter(words):
    result = ''
    for word in words:
        if word.isalpha():
            result += word
    return result[::-1]
````

# 편-안한 단어

> (QWERTY 키보드를 사용하여 타이핑을 한다고 가정할 때) '편안한 단어'는 타이핑 할 때 **손을 번갈아 칠 수 있는 단어**를 말합니다.
>
> 단어를 인자로 받아 그것이 '편안한 단어'인지 여부를 True/False로 반환하는 함수를 만드세요.(모든 단어는 a ~ z까지 오름차순으로 구성된 문자열입니다.)

> 문자 목록
- 왼손: q, w, e, r, t, a, s, s, d, f, g, z, x, c, v, b
- 오른손: y, u, i, o, p, h, j, k, l, n, m

````python
# 여기에 코드를 작성하세요.
````

````python
def comfortable_word(words):
    left_word, right_word = 'qwertasdfgzxcvb', 'yuiophjklnm'
    #왼손이면 True
    left = True if words[0] in left_word else False
    
    #2번째부터 비교 시작
    for word in words[1:]:
        #2번째 나온 단어가 왼손이고 앞 단어도 왼손이라면 False
        if word in left_word and left:
            return False
        #2번재 나온 단어가 오른손이고 앞 단어도 오른손이라면 False
        if word in right_word and not left:
            return False
        left = not left#왼손 오른손을 변경해서 반복하기 위해서
    return True
````

````python
def comfortable_word(words):
    left = ['q', 'w', 'e', 'r', 't', 'a', 's', 's', 'd', 'f', 'g', 'z', 'x', 'c', 'v', 'b']
    right = ['y', 'u', 'i', 'o', 'p', 'h', 'j', 'k', 'l', 'n', 'm']
    for num in range(0,len(words)-1):
        if words[num] in left and not words[num+1] in right:
                return False
        if words[num] in right and not words[num+1] in left:
                return False
    return True
````

````python
def comfortable_word(n):
    left = 'qwertasdfgzxcvb'
    right = 'yuiophjklnm'
    result = []
    for i in range(len(n)):
        if n[i] in left:
            result.append('l')
        elif n[i] in right:
            result.append('r')
    solution = ''.join(result)
    if 'll' in solution or 'rr' in solution:
        return False
    else:
        return True
````

````python
def pick_and_sum(words):
    #결과를 낼 result 정수 하나와 잠시 저장해 둘 num 문자열 하나
    result = 0
    num = ''
    for word in words:
        if word.isdecimal():
            num += word
        elif num: #num에 값이 있다면
            result += int('num')
            num = ''#num의 초기화
    return result
print(pick_and_sum('The30quick20brown10f0x1203jumps914ov3r1349the102l4zy dog'))
````

````python
def pick_and_sum(words):
    for word in words:
        if not word.isdecimal():
            words = words.replace(word, ' ')
#   print(words)
#   print(words.split())
    return map(int, words.split())
````
