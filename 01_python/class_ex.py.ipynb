# 클래스 연습해보기

```python
class Person:
    name = 'Daejeon'
```

```python
p1 = Person()#인스턴스 만들기: 클래스를 호출해서..
p2 = Person()
```

```python
print(p1.name)
```

```python
print(p2.name)
```

선언한 클래스 Person에는 멤버 변수(데이터)인 name만 포함되어있다. 이후에 생성된 p1, p2 인스턴스에는 아직 인스턴스 객체의 특화된 데이터가 없기 때문에 여전히 클래스 객체에서 데이터를 참조한다.

```python
p1.name = 'Seoul'
print(p1.name)
```

```python
print(p2.name)
```

인스턴스 객체 p1의 멤버 변수인 name을 'Seoul'로 변경하면 인스턴스 p1의 이름공간에 존재하는 name이라는 변수에 변경된 데이터를 저장.
(원래는 Person 클래스의 멤버 변수인 name(Daejeon)이라는 데이터를 참조함.)
하지만 인스턴스 객체 p2의 경우 멤버 변수인 name은 변경되지 않았기 때문에 Person 클래스의 멤버 변수 name을 참조한다.

```python
Person.title = 'Welcome Daejeon'
```

```python
print(p1.title)#title 만들기 전에 선언된 것이어도 참조 가능
print(p2.title)#p1, p2 인스턴스 모두에서 해당 멤버 변수(객체)에 접근 가능
```

```python
print(Person.title)
#클래스 객체에서도 접근 가능하다.
```

파이썬은 클래스 객체와 인스턴스 객체에 동적으로 멤버 변수를 추가/삭제할 수 있다.
클래스 객체 Person에 멤버 변수인 title을 추가하였다. 이렇게 추가된 변수는 인스턴스 객체를 통해 접근하려는 경우, 인스턴스 객체에는 같은 이름의 다른 멤버 변수가 없기 때문에 클래스 객체에 이름 공간을 참조해 정상적으로 접근할 수 있다.

```python
p1.water = '아리수'
print(p1.water)
```

```python
print(p2.water)#얘는 속성에러. Person에 워터라는 속성이 없음
```

- p1 인스턴스 객체에 water라는 멤버 변수를 추가하고, p2 인스턴스 객체를 통해 접근하는 경우에는 p2의 이름공간과 Person의 이름 공간에서는 해당 변수의 이름을 찾을 수 없기 때문에 에러가 발생한다.                                                                 
- 인스턴스 객체에 동적으로 멤버 변수를 추가한 경우에는 추가한 인스턴스 객체를 통해서만 접근할 수 있다.
    - p1.water는 p1 인스턴스 객체에만 접근한다.
    - Person.title은 클래스 객체에 접근하기 때문에 p1과 p2 인스턴스는 모두 title이라는 멤버 변수를 참조할 수 있다.
